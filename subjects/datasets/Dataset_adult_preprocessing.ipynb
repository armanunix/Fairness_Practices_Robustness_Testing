```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
le = LabelEncoder()
se = StandardScaler()
#train = pd.read_csv('train.csv')
#test = pd.read_csv('test.csv')
#df = pd.concat([train,test])
df =  pd.read_csv('adult_raw.csv')
# cat_columns = ['contact','job','marital','education','default','housing','loan','month','campaign','poutcome','y']

# columns = ['age', 'job', 'education', 'default', 'balance', 'housing',
#        'loan', 'contact', 'duration', 'pdays',
#         'poutcome', 'y']

# continous_col = ['balance', 'duration','pdays']

cat_columns = ['workclass','education','marital-status','occupation','relationship','race','gender','native-country','income']

columns = ['age', 'workclass',   'educational-num',
       'marital-status', 'occupation', 'relationship', 'race', 'gender',
        'hours-per-week', 'native-country',
       'income']

continous_col = ['fnlwgt', 'capital-gain','capital-loss','hours-per-week']
for col in cat_columns:
    df[col] = le.fit_transform(df[col])
    
for col in continous_col:
        df[col] = se.fit_transform(df[col].to_numpy().reshape(-1,1))    
df['age'] = pd.cut(df['age'],9, labels=[1, 2,3,4,5,6,7,8,9])
df = df[columns]
# col_list={}
# for col in columns[:-1]:
#     ind = columns.index(col)
#     df1 = df[columns[:ind]+columns[ind+1:]].copy()
#     acc=[]
#     f1=[]
#     for i in range(10):
#         model = LogisticRegression()
#         X = df1.to_numpy()[:,:-1]
#         y = df1.to_numpy()[:,-1]
#         X_train, X_text, y_train, y_test = train_test_split(X, y,train_size=0.8)
#         model.fit(X_train, y_train)
#         acc_temp = model.score(X_text, y_test)
#         pred = model.predict(X_text)
#         f1_temp = f1_score(y_test,pred)
#         acc.append(acc_temp)
#         f1.append(f1_temp)
#     col_list[col] = [np.mean(acc),np.mean(f1)]

#col_list 

model = LogisticRegression()
X = df.to_numpy()[:,:-1]
y = df.to_numpy()[:,-1]
X_train, X_text, y_train, y_test = train_test_split(X, y,train_size=0.8)
model.fit(X_train, y_train)
acc_temp = model.score(X_text, y_test)
pred = model.predict(X_text)
f1_temp = f1_score(y_test,pred)
print(acc_temp)
print(f1_temp)
```
