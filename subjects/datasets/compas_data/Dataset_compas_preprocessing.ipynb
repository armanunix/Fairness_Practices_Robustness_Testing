```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
le = LabelEncoder()
df = pd.read_csv('compas-scores-two-years.csv')

columns = ['sex','race','age_cat', 
           'priors_count','juv_fel_count','r_charge_degree','two_year_recid']


df['sex'] = le.fit_transform(df['sex'])
df['age_cat'] = le.fit_transform(df['age_cat'])
#df['race'] = le.fit_transform(df['race'])
df['race'] = df['race'].transform(lambda x: 1 if x=='Caucasian' else 0)
df['r_charge_degree'] = df['r_charge_degree'].transform(lambda x: '0' if str(x)=='nan' else x)
df['r_charge_degree'] = le.fit_transform(df['r_charge_degree'])
df = df[columns]
# df['juv_fel_count'] = df['juv_other_count'] + df['juv_misd_count'] + df['juv_fel_count'] 
# df = df.drop(columns=['juv_other_count','juv_misd_count'])
model = LogisticRegression()
X = df.to_numpy()[:,:-1]
y = df.to_numpy()[:,-1]
X_train, X_text, y_train, y_test = train_test_split(X, y,train_size=0.8)
model.fit(X_train, y_train)
print('Acc',model.score(X_text, y_test))
pred = model.predict(X_text)
print('F1',f1_score(y_test,pred))
```
